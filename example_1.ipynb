```python
import pandas as pd
import numpy as np
import csv
import json
```

```python
df = pd.read_csv('working_hours.csv')
```

```python
df = df.drop('Unnamed: 0',axis =1)
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
df_total_country = df.groupby('country_id').customer_id.nunique().reset_index()
```

```python
df_total_country
```

```python
df_hour_range = df.groupby(['country_id','day','range']).customer_id.nunique().reset_index()
```

```python
df_hour_range
```

```python
new_df = df_hour_range.merge(df_total_country,how = 'left', on = 'country_id').reset_index(drop = True)
```

```python
new_df
```

```python
new_df['Percentage'] = (new_df['customer_id_x'] / new_df['customer_id_y']) *100
```

```python
new_df
```

```python
new_df = new_df.rename(columns={'customer_id_x':'Customers_by_hour_range','customer_id_y':'Total_customers_by_country'})
```

```python
new_df
```

```python
new_df['Percentage']= new_df['Percentage'].round(decimals = 2)
```

```python
final_df = new_df[new_df['Percentage']>5].reset_index(drop = True)
```

```python
final_df
```

```python
final_df = final_df.reindex()
final_df = final_df.drop(labels = ['Customers_by_hour_range','Total_customers_by_country'], axis =1)
```

```python
final_df #ordenarlo alfabeticamente por pais y de maypor a menor por porcentaje
#En el mail acalrarle como calculamos el porcentaje y todos los criterios usados
```

```python
final_df = final_df.sort_values(['country_id','Percentage'],ascending = [True,False])
```

```python
final_df
```

```python
final_df.to_csv('hour_range_percentage.csv')
```
